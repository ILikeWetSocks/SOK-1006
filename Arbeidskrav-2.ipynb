## Arbeidskrav 2

#### *(i) Tegn begge etterspørselsfunskjonene i samme figur. Har du noe grunnlag for å si at den ene er "mer/mindre elastisk enn den andre"?* 

For å finne ut hvilke etterspørselsfunksjon som var mer elastisk så måtte jeg først finne ut elasiteten til de ulike funksjonene. Siden priselasiteten kan variere fra punkt til punkt så bestemte jeg meg for å regne ut priselasiteten i midtpunktet for hver etterspørselsfunksjon. Det blir *p = 75* for *Etterspørsel 1 (E1)* og *p = 50* for *etterspørsel 2(E2)*.

Matte for E1:

*$$\epsilon_1^D = \frac{dX^D}{X^D} * \frac{p}{X^D}$$*

*$$\epsilon_1^D = -5 * \frac{75}{125} = -3$$*

Matte for E2:

*$$\epsilon_2^D = -1 * \frac{50}{100} = -0.5$$*

*Her kan vi se at $\epsilon^D$ til  E1 > E2 siden -3 er mer elasitsk enn -0.5*

*En mye enklere måte for å se hvem som er mer elastisk er å se hvor raskt figuren stiger eller synker*


```python
#lager p-verdier
p_verdi = np.linspace(1, 160, 1000) 

#lager etterspørselsfunksjonene
def etterspørsel_1(p):
    return 500 - 5*p

def etterspørsel_2(p):
    return 150 - p

#funksjon for å lage figur
def create_ax():
    fig, ax = plt.subplots()
    ax.set_ylabel('Pris (p)', loc='top', rotation = 0)
    ax.set_xlabel('KMengde (x)', loc='right')
    ax.set(xlim=(0,160))
    ax.set(ylim=(0,550))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    return fig, ax

fig, ax = create_ax() #tegner figurene
ax.plot(p_verdi, etterspørsel_1(p_verdi), label = 'Etterspørsel 1')
ax.plot(p_verdi, etterspørsel_2(p_verdi), label='Etterspørsel 2')
ax.set_title('Etterspørsels funksjoner')
ax.legend(loc='best', frameon = False) #viser titler og fjerner rammen rundt dem
plt.show()
```

#### *(ii) Tegn to figurer med hver sin etterspørselsfunksjon, og tegn også tilbud $X^S=-20+1.5p$ på begge.* 

```python
def tilbud_1(p):
    return (-20) + 1.5*p

#figur 1
fig, ax = create_ax() #tegner figurene
ax.plot(p_verdi, etterspørsel_1(p_verdi), label = 'Etterspørsel 1')
ax.plot(p_verdi, tilbud_1(p_verdi), label='tilbud_1')
#tegner likevektspunkt
ax.hlines(y = 100, xmin=0, xmax=80, linestyles='dashed', label='Likevekt', color="black")
ax.vlines(x = 80, ymin=0, ymax=100, linestyles='dashed', color="black")

likevekt_tekst = "Likevekt er i (80, 100)".format(80, 100)
ax.text(100, 150, likevekt_tekst, ha='right', va='bottom', backgroundcolor='white', fontsize=10)

ax.set_title('Figur 1')
ax.legend(loc='best', frameon = False) #viser titler og fjerner rammen rundt dem
plt.show()

#figur 2 -------------------------------------------------------------------------------------------------------
fig, ax = create_ax() #tegner figurene
ax.plot(p_verdi, etterspørsel_2(p_verdi), label = 'Etterspørsel 2')
ax.plot(p_verdi, tilbud_1(p_verdi), label='tilbud_1')

ax.hlines(y = 82, xmin=0, xmax=68, linestyles='dashed', label='Likevekt', color="black")
ax.vlines(x = 68, ymin=0, ymax=82, linestyles='dashed', color="black")

likevekt_tekst = "Likevekt er i (68, 82)".format(80, 100)
ax.text(100, 150, likevekt_tekst, ha='right', va='bottom', backgroundcolor='white', fontsize=10)

ax.set_title('Figur 2')
ax.legend(loc='best', frameon = False) #viser titler og fjerner rammen rundt dem
plt.show()
```

#### *(iii) Anta nå at produksjonskostnader faller slik at tilbudet er nå gitt ved $X^S=-10+1.5p$. Tegn dette inn i begge figurene og undersøk effekten på likevektspris og -kvantum. Hvordan er dette påvirket at etterspørselens priselastisitet? Forklar intuisjonen bak dine funn.* 

```python
def tilbud_2(p):
    return (-10) + 1.5*p

#figur 3
fig, ax = create_ax() #tegner figurene
ax.plot(p_verdi, etterspørsel_1(p_verdi), label = 'Etterspørsel 1')
ax.plot(p_verdi, tilbud_2(p_verdi), label='tilbud_2')
#tegner likevektspunkt
ax.hlines(y = 107.7, xmin=0, xmax=78.5, linestyles='dashed', label='Likevekt', color="black")
ax.vlines(x = 78.5, ymin=0, ymax=107.7, linestyles='dashed', color="black")

likevekt_tekst = "Likevekt er i (78.5, 107.7)".format(80, 100)
ax.text(100, 150, likevekt_tekst, ha='right', va='bottom', backgroundcolor='white', fontsize=10)

ax.set_title('Figur 3')
ax.legend(loc='best', frameon = False) #viser titler og fjerner rammen rundt dem
plt.show()

#figur 4 -------------------------------------------------------------------------------------------------------
fig, ax = create_ax() #tegner figurene
ax.plot(p_verdi, etterspørsel_2(p_verdi), label = 'Etterspørsel 2')
ax.plot(p_verdi, tilbud_2(p_verdi), label='tilbud_2')

ax.hlines(y = 86, xmin=0, xmax=64, linestyles='dashed', label='Likevekt', color="black")
ax.vlines(x = 64, ymin=0, ymax=86, linestyles='dashed', color="black")

likevekt_tekst = "Likevekt er i (64, 86)".format(80, 100)
ax.text(100, 150, likevekt_tekst, ha='right', va='bottom', backgroundcolor='white', fontsize=10)

ax.set_title('Figur 4')
ax.legend(loc='best', frameon = False) #viser titler og fjerner rammen rundt dem
plt.show()
```

Jeg regner ut likevekten til de ulike figurene ved likevektspunktet

1. Likevekten til *Etterspørsel1* med *Tilbud1* er *-6.25*

2. Likevekten til *Etterspørsel1* med *Tilbud2* er *-6.86*

3. Likevekten til *Etterspørsel2* med *Tilbud1* er *-1.2*

4. Likevekten til *Etterspørsel2* med *Tilbud2* er *-1.34*

Her kan vi se at priselasiteten til Etterspørselsfunksjonene har økt med *Tilbud2* funksjonen i forhold til *Tilbud1* funksjonen i likevektspunktet.

Det vil si at en synking av produksjonskostnadder vil øke priselasiteten til etterspørselsfunksjonene.
