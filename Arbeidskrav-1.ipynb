### 6. Oppgave

Et marked har følgende invers tilbud og etterspørsel:
$$ p = A+Bx \;(invers \;tilbud)$$
$$ p = a - bx \;(invers \;etterspørsel).$$
(i) Her er a, b, A, B alle positive tall. Hva viser de?     
(ii) Regn ut markedets likevekt. Finn en betingelse som sikrer at det er positiv omsetning i markedet.     
(iii) Regn ut effekten som endringer i a, b, A, B har på likevekten.     
(iv) Sett in verdier for a, b, A, B og lag en figur som viser markedslikvekten.      
(v) Lag 4 figurer som viser viser effektene som du identifiserte i (iii). Bruk disse figurene til å gi en intuitiv forklaring på dine funn.


(i)

Som vi ser så er markedets tilbud og etterspørselsfunksjoner lineære funksjon. Det  først leddet av funksjonen med variabel $A$ eller $a$ viser hvor funkjsonen krysser y-aksen der $x=0$. Den andre leddet av funksjone &plusmn; $Bx$ eller &plusmn; $bx$ viser stigningen av funksjonen.

(ii)

Siden funksjonene bare er variabler så kan jeg bare finne svaret på en likevekt med formelen som kommer fra koden under og ikke et spesifikt punkt.

```python
#Definerer symboler og sørger for at integerne er positive
x, A, a, B, b = sp.symbols('x A a B b', positive = True)

#Lager fomrler
tilbud = A + B*x
etterspørsel = a - b*x

#Gjør dem om til formler med sympy
formler = sp.Eq(tilbud, etterspørsel)
#finner likevekt med sp.solve. Løser for x
likevekt = sp.solve(formler, x)

#Returnerer første svar og printer
x_punkt = likevekt[0]
print("Likevektspunktet for alle positive integer finnes når x =",x_punkt)
```

(iii)

som sagt tidligere så er funksjonene lineære funksjoner. Det betyr at det er veldig enkelt å se hvordan en endring av verdiene kan påvirke likevektens verdi

med formlene tilbud og etterspørsel så vil variablene $A$ og a bestemme startpunktet til grafen der linjen krysser $y-aksen$ der $x = 0$. Hvis $A$ er et stort tall så vil startpunktet til tilbud være stort. Derimot hvis $a$ har en stor verdi så vil etterspørselen ha en stor startverdi. $Bx$ og $bx$ forteller oss om hvor fort tilbud stiger($Bx$) og etterspørsel synker($bx$). Desto større verdier jo raskere vil vi få en likevekts verdi lagns $x-aksen$.

(iv)

Setter inn verdier på tidligere kode og tegner grafen.

```python

#Definerer verdier
A = 50
a = 200
B = 10
b = 15
x_verdier = np.linspace(1,10,100)
tilbud_verdier = A + B*x_verdier
etterspørsel_verdier = a - b*x_verdier

likevekt_x = (-A + a)/(B + b)
likevekt_y = A + B*likevekt_x

#plotter likevekt
plt.plot(x_verdier, tilbud_verdier, label = "Tilbud")
plt.plot(x_verdier, etterspørsel_verdier, label = "Etterspørsel")
plt.scatter(likevekt_x, likevekt_y, label = "Likevekt", color = "Green")

plt.legend() #legger til labels
plt.grid() 
plt.show()
```
